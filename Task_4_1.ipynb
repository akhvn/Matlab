```python
import math
import matplotlib.pyplot as plt

V_all=[]
v_all=[]
X_all=[]
```

```python
"""Исходные данные"""
m = 0.01 # в киллограммах
l = 0.1 # в метрах
g = 9.81 # м/c
tau = 0.001 #с (тау)
N = int(11/tau) #количество измерений (по тау)
```

```python
"""Начальные условия"""
x_0 = 0
v_0 = 20
v_all.append(v_0)
```

```python
def energy(x,v):
    return m*(l**2)*((v**2)/2 + g*(1-math.cos(x))/l)

def func(x):
    return -g*math.sin(x)/l

def draw(X,Y):
    plt.xlabel("x") 
    plt.ylabel("V")
    plt.grid()
    plt.minorticks_on()
    plt.grid(which='minor',color = 'black', linestyle = ':')
    plt.plot(X,Y)
```

# метод "предиктор-корректор"

```python
V = [v_0]
X = [x_0]
T = [0]
E = [energy(x_0,v_0)]
for i in range(N):
    v = V[-1]
    x = X[-1]
    t = T[-1]
    v_= v + tau * func(x)
    x_= x + tau * v
    V.append(v + tau * (func(x) + func(x_)) / 2)
    X.append(x + tau * (v + v_) / 2)
    T.append(t + tau)
    #E.append(energy(X[-1],V[-1]))
draw(X,V)
V_all.append(V)
X_all.append(X)
```
